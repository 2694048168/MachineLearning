```python
# Accessing and manipulating pixels in OpenCV with BGR images

import cv2 as cv

img = cv.imread("./logo.png")

# height × width × channels
dimensions = img.shape
print(dimensions)
# img.size is equal to the multiplication of height × width × channels
total_num_elements = img.size
print(total_num_elements)
# data type
image_dtype = img.dtype
print(image_dtype)
cv.imshow("Original image", img)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
# (x=40, y=6)
(b, g, r) = img[12, 120]
print(b, g, g)

b = img[12, 120, 0]
print(b)

# The pixel values can be also modified in the same way - (b, g, r) format:
img[12, 120] = (0, 0, 255)

# In this case, we get the top left corner of the image:
top_left_corner = img[0:150, 0:150]
cv.imshow("top_left_corner",top_left_corner)
cv.waitKey()
cv.destroyAllWindows()
```

```python
# Accessing and manipulating pixels in OpenCV with grayscale images

gray_img = cv.imread("./logo.png", cv.IMREAD_GRAYSCALE)
# height × width
dimensions = gray_img.shape
print(dimensions)
# img.size is equal to the multiplication of height × width
total_num_elements = gray_img.size
print(total_num_elements)
# data type
image_dtype = gray_img.dtype
print(image_dtype)
cv.imshow("Original GRAY image", gray_img)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
# BGR order in OpenCV
# RGB order in matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

# method 1 to convert
# Load image using cv2.imread:
img_OpenCV = cv.imread('logo.png')
# Split the loaded image into its three channels (b, g, r):
b, g, r = cv.split(img_OpenCV)
# Merge again the three channels but in the RGB format:
img_matplotlib = cv.merge([r, g, b])

# Show both images (img_OpenCV and img_matplotlib) using matplotlib
# This will show the image in wrong color:
plt.subplot(121)
plt.imshow(img_OpenCV)
# This will show the image in true color:
plt.subplot(122)
plt.imshow(img_matplotlib)
plt.show()
```

```python
import numpy as np

# To stack horizontally (img_OpenCV to the left of img_matplotlib):
img_concats = np.concatenate((img_OpenCV, img_matplotlib), axis=1)
# Now, we show the concatenated image:
cv.imshow('bgr image and rgb image', img_concats)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
# One consideration to take into account is that cv2.split() is a time-consuming operation.
# method 2 to convert
# Transform the image BGR to RGB using Numpy capabilities:
img_matplotlib = img_OpenCV[:, :, ::-1]
# Show both images (img_OpenCV and img_matplotlib) using matplotlib
# This will show the image in wrong color:
plt.subplot(121)
plt.imshow(img_OpenCV)
# This will show the image in true color:
plt.subplot(122)
plt.imshow(img_matplotlib)
plt.show()
```
