## Introduction to Flask

```python
# hello.py

# Import required packages:
from flask import Flask

app = Flask(__name__)
@app.route("/")
def hello():
  return "Hello World!"

if __name__ == "__main__":
  app.run()
```

```python
# hello_external.py

# Import required packages:
from flask import Flask

app = Flask(__name__)
@app.route("/")
def hello():
  return "Hello World!"

if __name__ == "__main__":
  # Add parameter host='0.0.0.0' to run on your machines IP address:
  # http://192.168.1.100:5000/
  app.run(host='0.0.0.0')
```
