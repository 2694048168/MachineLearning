```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
#
# @author: weili
# @function: 绘制不同均值和方差的高斯函数曲线

import math
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def normal_distribution(x, mean, sigma):
    return np.exp(-1*((x-mean)**2)/(2*(sigma**2)))/(math.sqrt(2*np.pi) * sigma)

mean1, sigma1 = 0, 1
x1 = np.linspace(mean1 - 6*sigma1, mean1 + 6*sigma1, 100)

mean2, sigma2 = 0, 2
x2 = np.linspace(mean2 - 6*sigma2, mean2 + 6*sigma2, 100)

# mean5, sigma5 = -5, 2
# x5 = np.linspace(mean5 - 6*sigma5, mean5 + 6*sigma5, 100)

mean3, sigma3 = 5, 1
x3 = np.linspace(mean3 - 6*sigma3, mean3 + 6*sigma3, 100)

mean4, sigma4 = -5, 1
x4 = np.linspace(mean4 - 6*sigma4, mean4 + 6*sigma4, 100)

y1 = normal_distribution(x1, mean1, sigma1)
y2 = normal_distribution(x2, mean2, sigma2)
y3 = normal_distribution(x3, mean3, sigma3)
y4 = normal_distribution(x4, mean4, sigma4)
# y5 = normal_distribution(x5, mean5, sigma5)

fig = plt.figure(figsize = (10,8), dpi = 120) 
plt.plot(x1, y1, 'r', label='m=0,sig=1')
plt.plot(x2, y2, 'g', label='m=0,sig=2')
plt.plot(x3, y3, 'b', label='m=5,sig=1')
plt.plot(x4, y4, 'm', label='m=-5,sig=1')
# plt.plot(x5, y5, 'y', label='m=0,sig=5')
plt.title(r"Gaussian Function $ G(x) = \frac{1}{\sqrt{2 \pi} \sigma} exp(- \frac{(x - \mu)^2}{2 \sigma^2}) $")
plt.legend()
plt.grid()
plt.savefig("./Gaussian.png", bbox_inches='tight')
plt.show()
```

```python
import numpy as np
import random
import cv2 as cv
from matplotlib import pyplot as plt
%matplotlib inline

def sp_noise(image,prob):
    '''
    添加椒盐噪声
    prob:噪声比例
    '''
    output = np.zeros(image.shape, np.uint8)
    thres = 1 - prob
    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
            rdn = random.random()
            if rdn < prob:
                output[i][j] = 0
            elif rdn > thres:
                output[i][j] = 255
            else:
                output[i][j] = image[i][j]
    return output

def gasuss_noise(image, mean=0, var=0.001):
    '''
        添加高斯噪声
        mean : 均值
        var : 方差
    '''
    image = np.array(image/255, dtype=float)
    noise = np.random.normal(mean, var ** 0.5, image.shape)
    out = image + noise
    if out.min() < 0:
        low_clip = -1.
    else:
        low_clip = 0.
    out = np.clip(out, low_clip, 1.0)
    out = np.uint8(out*255)
    #cv.imshow("gasuss", out)
    return out

# Read image
img = cv.imread("./lena.jpg", 0)

# 添加椒盐噪声，噪声比例为 0.02
out1 = sp_noise(img, prob=0.02)

# 添加高斯噪声，均值为0，方差为0.001
out2 = gasuss_noise(img, mean=0, var=0.001)

# cv.imshow("Original", img)
# cv.imshow("Add Salt and Pepper noise", out1)
# cv.imshow("Add Gaussian noise", out2)
# cv.waitKey(0)
# cv.destroyAllWindows()

# 实现在 notebook 中显示图像结果，而不是重新打开一个窗口
def imshow_notebook(name, img):
    # OpenCV 颜色通道顺序 BGR
    # Matplotlib 颜色通道顺序 RGB
    img = img[:, :, (2, 1, 0)]
    plt.imshow(img)
    plt.title(name)
    plt.show()
    
# 实现在 notebook 中显示灰度图像，而不是重新打开一个窗口
def imshow_gray_notebook(name, img):
    # OpenCV 颜色通道顺序 BGR
    # Matplotlib 颜色通道顺序 RGB
    img = img[:, :]
    plt.imshow(img, cmap="gray")
    plt.title(name)
    plt.show()

# 显示图像
fig = plt.figure(figsize = (10,8), dpi = 120) 
# plt.figure(1)
# plt.subplot(131)
plt.axis('off')  # 关闭坐标轴
# plt.title('Original')
# plt.imshow(img)
# plt.savefig("./Original.png")
cv.imwrite("./Original.png", img)
imshow_gray_notebook("Original", img)
# plt.subplot(132)

fig = plt.figure(figsize = (10,8), dpi = 120) 
plt.axis('off')
# plt.title('Add Salt and Pepper noise')
# plt.imshow(out1)
# plt.savefig("./Salt_and_Pepper_Noise.png")
cv.imwrite("./Salt_and_Pepper_Noise.png", out1)
imshow_gray_notebook("Add Salt and Pepper noise", out1)
# plt.subplot(133)

fig = plt.figure(figsize = (10,8), dpi = 120) 
plt.axis('off')
# plt.title('Add Gaussian noise')
# plt.imshow(out2)
# plt.savefig("./Gaussian_Noise.png")
cv.imwrite("./Gaussian_Noise.png", out2)
imshow_gray_notebook("Add Gaussian noise", out2)
plt.show()
```

```python
# 椒盐（脉冲）噪声消除
#encoding: utf-8
import math
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def medium_filter(im, x, y, step):
    sum_s=[]
    for k in range(-int(step/2),int(step/2)+1):
        for m in range(-int(step/2),int(step/2)+1):
            sum_s.append(im[x+k][y+m])
    sum_s.sort()
    return sum_s[(int(step*step/2)+1)]

def median_filter(image, win=3):
    H, W, C = image.shape
    result = image.copy()
    for h in range(1, H-2):
        for w in range(1, W-2):
            for c in range(C):
                result[h, w, c] = np.median(result[h:h+win, w:w+win, c])
    return result

def _mean_filter(im, x, y, step):
    sum_s = 0
    for k in range(-int(step/2),int(step/2)+1):
        for m in range(-int(step/2),int(step/2)+1):
            sum_s += im[x+k][y+m] / (step*step)
    return sum_s

def mean_filter(image):
    K = ([1, 1, 1],
         [1, 1, 1],
         [1, 1, 1])
    K = np.array(K)
    H, W, C = image.shape
    result = image.copy()
    # 因为卷积核是以左上角为定位，所以遍历时最后要停到H-2处
    for h in range(1, H-2):
        for w in range(1, W-2):
            for c in range(C):
                result[h, w, c] = sum(sum(K * result[h:h+K.shape[0], w:w+K.shape[1], c])) // 9
    return result

if __name__ == '__main__':
    img = cv.imread("./Original.png")
    result_img = median_filter(img)
    fig = plt.figure(figsize = (10,8), dpi = 120) 
    plt.axis('off')  # 关闭坐标轴
    cv.imwrite("./result_denoise_salt.png", result_img)
    imshow_gray_notebook("result_img", result_img)
    
    result_img2 = mean_filter(img)
    fig = plt.figure(figsize = (10,8), dpi = 120) 
    plt.axis('off')  # 关闭坐标轴
    cv.imwrite("./mean_denoise_salt.png", result_img)
    imshow_gray_notebook("result_img2", result_img2)
```

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Gaussian filter
def gaussian_filter(img, K_size=3, sigma=4):
    if len(img.shape) == 3:
        H, W, C = img.shape
    else:
        img = np.expand_dims(img, axis=-1)
        H, W, C = img.shape
    ## Zero padding
    pad = K_size // 2
    out = np.zeros((H + pad * 2, W + pad * 2, C), dtype=np.float)
    out[pad: pad + H, pad: pad + W] = img.copy().astype(np.float)
    ## prepare Kernel
    K = np.zeros((K_size, K_size), dtype=np.float)
    for x in range(-pad, -pad + K_size):
        for y in range(-pad, -pad + K_size):
            K[y + pad, x + pad] = np.exp( -(x ** 2 + y ** 2) / (2 * (sigma ** 2)))
    K /= (2 * np.pi * sigma * sigma)
    K /= K.sum()
    tmp = out.copy()
    # filtering
    for y in range(H):
        for x in range(W):
            for c in range(C):
                out[pad + y, pad + x, c] = np.sum(K * tmp[y: y + K_size, x: x + K_size, c])
    out = np.clip(out, 0, 255)
    out = out[pad: pad + H, pad: pad + W].astype(np.uint8)
    return out

if __name__ == '__main__':
    img = cv.imread("./Gaussian_Noise.png")    
    gaussian_filter_img = gaussian_filter(img)
    fig = plt.figure(figsize = (10,8), dpi = 120) 
    plt.axis('off')  # 关闭坐标轴
    cv.imwrite("./gaussian_filter_img.png", gaussian_filter_img)
    imshow_gray_notebook("gaussian_filter_img", gaussian_filter_img)
```

二维高斯函数 $ G(x,y) = \frac{1}{(\sqrt{2 \pi} \sigma)^2} exp(- \frac{(x - \mu_x)^2 + (y - \mu_y)^2}{2 \sigma^2}) $

卷积核的半窗宽度设置为 $3\sigma$

根据经验法则，则卷积核的大小 $2 \times 3\sigma +  1$

平均数 $\mu = \frac{\sum\limits_{i=1}^{n}x_i}{n}$

总体方差 $\sigma^2 = \frac{\sum \limits_{i=1}^{n}(x_i - \mu)}{n}$
