## Matplotlib

```python
#!/usr/bin/env python3
# -*- coding: UTF-8 -*-
# @author
# @date2020-02-05  18:56
# @introductionmatplotlibpyplot
# @tips

# 
from matplotlib import pyplot as plt
from matplotlib import font_manager

# 
x = range(2,27,2)
y = [15,13,14,5,17,20,25,26,26,27,22,18,15]
y2 = [13,12,17,16,15,18,23,24,27,28,22,19,25]

# ,dpi
fig = plt.figure(figsize = (10,7), dpi = 80)

# matplotlib
# font_managerfname
# fc()fc-list :lang=zh
# fontproperties = my_font
my_font = font_manager.FontProperties( fname = r"C:/Windows/fonts/msyhbd.ttf" )

# +
plt.xticks(range(min(x),max(x)+1))
plt.yticks(range(1,28,1))

# 
plt.xlabel("", fontproperties = my_font)
plt.ylabel("", fontproperties = my_font)
#figtextxy
plt.figtext(0.9, 0.05, "", fontproperties = my_font)
plt.figtext(0.1, 0.9, "", fontproperties = my_font)
plt.title("", fontproperties = my_font)

# 
plt.grid(alpha=0.4)

# 
plt.text(0.95, 0.5, "", 
         fontsize=80, color='gray',
         ha='right', va='bottom', alpha=0.5, fontproperties = my_font)

# ,
plt.plot(x,y,label="",color="red")
plt.plot(x,y2,label="",color="green")

# loc0
plt.legend(prop = my_font,loc=0)

# svgpdfepspng
plt.savefig("./pyplot.svg")
plt.savefig("./pyplot.png")
plt.savefig("./pyplot.pdf")

# 
plt.show()
```
