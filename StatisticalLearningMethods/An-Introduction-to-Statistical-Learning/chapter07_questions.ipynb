### (a) Perform polynomial regression to predict wage using age. Use cross-validation to select the optimal degree d for the polynomial. What degree was chosen, and how does this compare to the results of hypothesis testing using ANOVA? Make a plot of the resulting polynomial fit to the data.




```python
# a function which performs the polynomial regression, fits and finds a cross validation error for a model with degree 
# from 1 to power
def poly_regression_cv(X,y,power):
  scores_dict = {}
  for n in range(1,power+1):
    poly = PolynomialFeatures(n,include_bias=False)
    X_transformed = poly.fit_transform(X.to_frame())
    lin_reg = LinearRegression()
    scores_dict[n] = (-np.mean(cross_val_score(lin_reg,X_transformed,y,scoring = 'neg_mean_squared_error',cv = 5)))
  return scores_dict    
```

```python
X = wage['age']
y = wage['wage']

scores_dict = poly_regression_cv(X,y,15)
```

### (b) Fit a step function to predict wage using age, and perform crossvalidation to choose the optimal number of cuts. Make a plot of the fit obtained

```python
def step_functions_cv(X,y,total_steps):    
  scores_dict = {}
  for cut in range(2,total_steps):    
    X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.5,random_state = 0)
    X_cut_train = pd.cut(X_train,cut)
    X_dummies_train = pd.get_dummies(X_cut_train)

    X_cut_test = pd.cut(X_test,cut)
    X_dummies_test = pd.get_dummies(X_cut_test)

    model = sm.OLS(y_train,X_dummies_train).fit()
    preds = model.predict(X_dummies_test)
    scores_dict[cut] = mean_squared_error(preds,y_test)
  return scores_dict
```

```python
scores_dict_steps = step_functions_cv(wage['age'],wage['wage'],16)
```

```python
plt.plot(list(scores_dict_steps.keys()),list(scores_dict_steps.values()))
plt.xlabel('Number of cuts')
plt.ylabel('Validation error')
```

#### we can see from the above graph that the best performance is when cuts = 8

 ### a) Split the data into a training set and a test set. Using out-of-state tuition as the response and the other variables as the predictors, perform forward stepwise selection on the training set in order to identify a satisfactory model that uses just a subset of the predictors.





```python
# taken from lab chapter 6
def forward_stepwise_selection(data,target):
  total_features = [[]]
  score_dict = {}
  remaining_features = [col for col in data.columns if not col == target]
  for i in range(1,len(data.columns)):
    best_score = 0;best_feature = None
    for feature in remaining_features:
      X = total_features[i-1] + [feature]
      model = LinearRegression().fit(data[X],data[target])
      score = r2_score(data[target],model.predict(data[X]))
      # print('For len {}, feature - {}, score is {}'.format(i,feature,score))

      if score > best_score:
        best_score = score
        best_feature = feature
    total_features.append(total_features[i-1] + [best_feature])
    remaining_features.remove(best_feature)
    score_dict[i] = best_score
  return total_features,score_dict
```

```python
total_features_fwd,score_dict_fwd = forward_stepwise_selection(college,'Outstate')
```

```python
# lets print the selceted features for first 5 models
for i in range(1,len(total_features_fwd)-1):
  print('The best model with {} features - {}'.format(i,total_features_fwd[i]))
  print('R_2 score is ',score_dict_fwd[i])
  print()
```

Y − βˆ1X1 = β0 + β2X2 + error

```python
lr1 = LinearRegression().fit(X1.reshape(-1,1),y - beta_hat[2]*X2)
beta_hat[1] = lr1.coef_
beta_hat[0] = lr1.intercept_
print(lr1.coef_)
print(lr1.intercept_)
```

In a toy example with p = 100, show that one can approximate the multiple linear regression coefficient estimates by repeatedly performing simple linear regression in a backfitting procedure. How many backfitting iterations are required in order to obtain a “good” approximation to the multiple regression coefficient estimates? Create a plot to justify your answer.







in this case , n = 100 and p = 100

```python
np.random.seed(1) # setting the seed so that we get the same number everytime

# define the X matrix, it is from 0 ot 1 ,and shape is (n,p) ~ (100,100)
X = np.random.rand(100,100)

# define the true weights, 
true_beta = np.random.uniform(-2.5,8.0,(100,1))

y = np.dot(X,true_beta) + np.random.randn(100,1)
```

```python
from tqdm import tqdm
diffs = []
betas_estimated = np.ones((100,1))
mlr_betas = LinearRegression().fit(X,y).coef_[0]
# one iteration of backfitting 
for _ in tqdm(range(1000)):
  for predictor in range(100):
    X_current = X[:,predictor]
#         print(X_current.shape)
    X_remaining = np.delete(X,predictor,axis = 1)
#         print(X_remaining.shape)
    weights_remaining = np.delete(betas_estimated,predictor,axis = 0)
#         print(betas_estimated.shape)
    betas_estimated[predictor] = LinearRegression().fit(X_current.reshape(-1,1),y - np.dot(X_remaining,weights_remaining)).coef_[0]
  diffs.append(np.sum(np.abs(betas_estimated - true_beta)))
```

for each iteration we will plot the sum difference between the coeff given by backtracking and by multiple linear regression

```python
plt.plot(np.arange(1,1001),diffs)
plt.ylabel('Sum of differnce')
plt.xlabel('Iterations')
```

From the graph above, we can say that the sum of difference decreases first, reaches a minimum and then increasesa again, So, we can say that the coefficients first goes to the optimal values and than they kind of overfit the data
