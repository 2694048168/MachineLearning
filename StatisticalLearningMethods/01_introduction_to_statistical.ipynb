## 1.1 统计学习
- 统计学习( **statistical learning** ) 是关于计算机基于数据构建概率统计模型并运用模型对数据进行预测与分析的学科

- 统计学习也称之为统计机器学习( **statistical machine learning** )，人们提及的机器学习，往往都是指统计机器学习

- **概率论 + 统计学 + 信息论 + 计算理论 + 最优化理论 + 计算机科学** 的交叉学科

- 计算机和网络为平台；数据为研究对象；构建并应用模型对数据进行预测与分析

- 数据( data ) ，提取数据特征——>抽象数据的模型——>发现数据的规律知识——>对数据进行分析与预测

- 数据：数字、文字、图像、视频、音频

- 统计学习关于数据的**基本假设**是同类数据具有一定的统计规律性，这是统计学习的前提

- 同类数据指的是具有某种共同性质的数据，具有统计规律性，可用概率统计方法进行处理，[ 随机变量描述数据中的特征；概率分布描述数据的统计规律 ]

- 统计学习中，以变量或者变量组表示数据，数据分为连续变量和**离散变量**，对数据还需要观测和收集流程

- 统计学习目的，对数据进行预测与分析，计算机智能化，计算机性能提高，发现新的知识与规律

- 通过构建概率统计模型实现目的，主要考虑学习什么样的模型和如何学习模型，使得模型能够对数据进行准确的预测与分析，同时考虑提高学习效率


- 统计学习方法：
    1. 给定有限用于学习的训练数据集合 **train_set**，用于测试的测试数据集 **test_set & validation_set**，假设数据是独立同分布产生的 IID
    2. 假设要学习的模型属于某一个函数的集合，称为假设空间 **hypothesis space**
    3. 应用某个评价准则 **evaluation criterion**, 从假设空间中选取一个最优的模型，使得其对于已知的训练数据和未知的测试数据在给定的评价准则下达到最优
    4. 最优模型的选取由算法实现，统计学习方法包含模型的假设空间、模型的选择的准则、模型的学习的算法，简称 **模型model、策略strategy、算法algorithm** 三要素


- 统计学习的研究
    1. 统计学习方法 **statistical learning method**, 旨在开发新的学习方法
    2. 统计学习理论 **statistical learning theory**, 研究方法的有效性与效率，统计学的基本理论
    3. 统计学习应用 **application of statistical learning**, 应用方法解决实际问题


- 统计学习的重要性
    1. 处理海量数据，大数据时代
    2. 计算机智能化
    3. 计算机科学的重要部分

## 1.2 监督学习
- **supervised learning**

- **input space** 输入空间：输入的所有可能取值的集合

- **output space** 输出空间：输出的所有可能取值的集合

- 每一个具体的输入是一个实例 **instance**，通常有特征向量 **feature vector** 表示

- **feature space** 特征空间：一个输入的所有特征向量存在的空间

- **sample** 样本或者样本点：输入与输出对

- 回归问题：输入变量和输出变量都是连续变量的预测问题

- 分类问题：输出变量为有限个离散变量的预测问题

- 标注问题：输入变量与输出变量都是变量序列的预测问题

- 统计学习假设数据具有一定的统计规律

- 监督学习的模型可以是概率模型或者非概率模型，有条件概率分布 $P(Y|X)$ 或者决策函数 decision function $Y = f(X)$, 对应具体的输出预测，记号为$P(y|x)$ 或者 $y = f(x)$

- 一般向量都是列向量，输入实例 x 的特征向量记号为: $ x = (x^1, x^2, ... , x^i, ..., x^n)^T $

- 一般使用表示多个输入变量中的第 i 个记号为：$ x_i = (x_i^1, x_i^2, ..., x_i^n)^T $

- 训练数据集一般表示为：$T = \{(x_1,y_1), (x_2,y_2), ..., (x_i,y_i), ..., (x_n,y_n)\}$

- 监督学习假设输入与输出的随机变量 X 和 Y 符合联合概率分布 P(Y|X)

- 训练数据与测试数据被认为是依联合概率分布 P(Y|X) 独立同分布产生的

## 1.3 统计学习三要素


- 模型:
    1. **hypothesis space** $\digamma$, 定义为决策函数的集合：$\digamma = \{f|Y = f(X)\}$, 其中 X 和 Y 是定义在输入空间和输出空间 $\chi$ 和 $Y$ 上的变量
    
    2. $\digamma$ 是一个参数向量决定的函数族：$\digamma = \{f|Y = f_\theta(X), \theta \in R^n \}$, 参数向量 $\theta$ 取值于 n 维欧氏空间$R^n$, 称之为参数空间 parameter space
    
    3. **hypothesis space** $\digamma = \{P|P(Y|X)\}$, $\digamma$ 是一个参数向量决定的条件概率分布族：$\digamma = \{P|P_\theta(Y|X), \theta \in R^n \}$


- 策略：
    1. 损失函数 **loss function** 或者 代价函数 **cost function** 度量预测错误的程度，度量模型一次预测的好坏
    
    2. 风险函数度量平均意义下模型预测的好坏
    3. 统计学习的损失函数：
        - **0-1 loss function** : $L(Y,f(X)) = \left\{\begin{array}{rl}
                                                1, & \text{if } Y \neq f(X) \\
                                                0, & \text{if } Y = f(X) \\
                                                \end{array} \right.$
        
        - **quadratic loss function** : $L(Y,f(X)) = (Y - f(X))^2$
        
        - **absolute loss function** : $L(Y,f(X)) = \vert Y - f(X)\vert$
        
        - **logarithmic loss function** or **log-likelihood loss function** : $L(Y,P(Y|X)) = -log P(Y|X)$
    4. 由于模型的输入和输出都是(X,Y)是随机变量，并且符合联合概率分布 $P(Y|X)$，则损失函数的期望：$R_{exp}(f) = E_p[L(Y,P(Y|X))] = \int_{\chi Y} L(y,f(x))P(x,y) \mathrm{d}x\mathrm{d}y$ 
    
    这是理论上模型f(X) 关于联合分布 P(Y|X) 的平均意义下的损失，称之为风险函数 **risk function** 或者期望损失 **expected loss**
    
    5. 由于 P(Y|X) 联合分布未知(如果已知道，通过数学计算就完成了)，$R_{exp}(f)$ 不能直接计算；
    
    一方面根据期望风险最小学习模型要用到联合分布，另一方面联合概率分布未知的，变态问题 **ill-formed problem**
    
    6. **train set** $T = \{(x_1,y_1), (x_2,y_2), \dots, (x_i,y_i), \dots, (x_N,y_N)\}$ 
    
    模型 f(X) 关于训练集的平均损失称之为经验风险 **empirical risk** 或者经验损失 **empirical loss**, 记号为 $R_{emp}$:
    
    $R_{emp} (f) = \frac{1}{N} \sum \limits_{i=1}^N L(y_i,f(x_i))$
    
    7. 期望风险 $R_{exp}(f)$ 是模型关于联合分布的期望损失；经验风险 $R_{emp}(f)$ 是模型关于训练集的平均损失；根据大数定理，当样本容量 N 趋于无穷大时，经验风险 $R_{emp}(f)$ 趋于期望风险 $R_{exp}(f)$
    
    可用经验风险来估计期望风险，但是实际情况训练集数据有限，所以估计很不理想，需要进行矫正；矫正基本策略：经验风险最小化和结构风险最小化
    
    8. 经验风险最小化 **empirical risk minimization, ERM** 的策略认为，经验风险最小的模型就是最优的模型，按照经验风险最小化求解最优模型就是求解最优化问题
    
    $\min \limits_{f \in \digamma} R_{emp}(f) = \min \limits_{f \in \digamma} \frac{1}{N} \sum \limits_{i=1}^N L(y_i,f(x_i))$, if 样本容量 N 足够大
    
    极大似然估计 **maximum likelihood estimation，MLE** 就是经验风险最小化的实例；当模型是条件概率分布，损失函数是对数损失函数时。
    
    9. 结构风险最小化 **structural risk minimization, SRM** 的策略认为，结构风险最小化的模型就是最优的模型，求解最优化问题
    
    $\min \limits_{f \in \digamma} R_{srm}(f) = \min \limits_{f \in \digamma} \frac{1}{N} \sum \limits_{i=1}^N L(y_i,f(x_i)) + \lambda J(f)$
    
    结构风险最小化为了防止过拟合 **over-fitting** 提出的策略，结构风险最小化等价于正则化 **regularization** ,结构风险在经验风险上加入表示模型复杂度的正则化项 **regularizer** 或者 惩罚项 **penalty term**
    
    **the definition of structural risk minimization** $R_{srm}(f) = \frac{1}{N} \sum \limits_{i=1}^N L(y_i,f(x_i)) + \lambda J(f)$
    
    其中 $J(f)$ 为模型的复杂度，是定义在假设空间 $\digamma$ 上的泛函；$\lambda \geq 0$ 是系数，用于权衡经验风险和模型复杂度
    
    贝叶斯估计中的最大后验概率估计 **maximum posterior probability estimation, MAP** 就是结构风险最小化的实例；当模型是条件概率分布，损失函数是对数损失函数，模型复杂度由模型的先验概率表示时。
    
  
- 算法：
    1. 算法指的是学习模型的具体计算方法，考虑用什么样的计算方法求解最优模型
    2. 统计学习问题归结为最优化问题，如果最优化问题有**显式的解析解**，那最优化就简单；如果没有解析解，需要用**数值计算**的方法求解
    3. 如何找到全局最优解，也就是**凸优化**的问题，并使得求解过程非常**高效**
    4. 利用已有的最优化方法；开发独自的最优化计算方法(算法)
    

## 1.4 模型评估与模型选择

- 训练误差 **training error** 就是模型关于训练数据的平均损失 $R_{emp} (f) = \frac{1}{N} \sum \limits_{i=1}^N L(y_i,f(x_i))$

- 测试误差 **testing error** 就是模型关于测试数据的平均损失

- 泛化误差 **generalization error** 就是学习方法对未知数据的预测能力

- 当假设空间含有不同复杂度(不同的参数个数/参数向量不同)的模型时，就面临着模型选择问题 **model selection**

- 一味的提高对训练数据的预测能力，所选择的模型复杂度往往比真模型更高，出现过拟合现象 **over-fitting**，模型的选择旨在避免过拟合并提高模型的预测能力

## 1.5 正则化与交叉验证

1. 正则化
    - 模型选择的典型方法就是正则化 **regularization**，正则项一般都是模型复杂度的单调递增函数，模型越复杂，正则化值就越大；正则化项可以是模型参数向量的范数

    - 正则化的作用就是选择经验风险与模型复杂度同时较小的模型，正则化符合奥卡姆剃刀 **Occam's razor**原理
    
        $L(w) = \frac{1}{N} \sum \limits_{i=1}^N (f(x_i;w) - y_i)^2 + \frac{\lambda}{2} \|w\|_2$
    
        $L(w) = \frac{1}{N} \sum \limits_{i=1}^N (f(x_i;w) - y_i)^2 + \lambda \|w\|$

    - 奥卡姆剃刀原理应用于模型选择：在所有可能选择的模型中，能够很好地解释已知数据并且十分简单才是最好的模型

    - 从贝叶斯估计的角度来看，正则化项对应模型的先验概率：假设复杂模型有较小的先验概率；简单模型有较大的先验概率

2. 交叉验证
    - 模型选择的方法还有交叉验证 **cross validation**
    
    - 样本数据充足，进行模型选择的简单方法就是随机将数据划分为：训练集**train set**、验证集**validation set**、测试集**test set**
      
      训练集用来训练模型；验证集用来模型的选择；测试集用来对最终的学习方法进行评估
      
    - 实际中样本数据是不充足的，采用交叉验证方法，重复使用数据
    
    - 简单交叉验证：数据集划分训练集和测试集；使用训练集来训练不同复杂度(参数向量不同)的模型；使用测试集来评价每一个模型的测试误差，选择测试误差最小的模型
    
    - S 折交叉验证**S-fold cross validation**：将数据集随机划分为 S 个互不相交的大小相同的子集；利用 S-1 个进行训练，1 个用于测试模型；重复对 S 中选择进行操作，选择 S 次平均测试误差最小的模型
    
    - 留一交叉验证**leave-one-out cross validation**：S = N；这种情况出现在数据容量很小很小
    

## 1.6 泛化能力

- 学习方法的泛化能力**generalization ability**是指该方法学习到的模型对未知数据的预测能力

- 通过测试误差来评价学习方法的泛化能力，依赖于测试数据集，该评价结果并不可靠

- 泛化误差**generalization error**定义为学到的模型对未知数据预测的误差

- 实质上泛化误差就是所学到模型的期望风险：$R_{exp}(f) = E_p[L(Y,P(Y|X))] = \int_{\chi Y} L(y,f(x))P(x,y) \mathrm{d}x\mathrm{d}y$

- 泛化误差上界**generalization error bound**：学习方法的泛化能力分析通过研究泛化误差的概率上界进行的
    
    泛化误差 $R_{exp}(f) \leq R_{emp}(f) + \varepsilon(d,N,\delta) $ 泛化误差上界
    
    其中 $\varepsilon(d,N,\delta) = \sqrt{\frac{1}{2N}(log d + log \frac{1}{\delta})}$
    
    其中 N 为样本容量；d 为假设空间的有限个函数个数；任意函数以概率 $1 - \delta$ 

## 1.7 生成模型与判别模型

1. 生成方法 **generative approach**，通过该方法所学到的模型称之为生成模型 **generative model**: 朴素贝叶斯、隐马尔科夫模型
    - 生成方法由数据学习联合概率分布 P(X,Y)，然后求出条件概率分布 P(Y|X)作为预测的模型，即生成模型
    
    $P(Y|X) = \frac{P(X,Y)}{P(X)}$
    
    之所有称之为生成方法，是因为模型表示了给定输入 X 产生输出 Y 的生成关系

2. 判别方法 **discriminative approach**，通过该方法所学到的模型称之为判别模型 **discriminative model**：K 近邻、感知机、决策树、逻辑斯蒂回归模型、最大熵模型、支持向量机、提升方法、条件随机场
    - 判别方法由数据直接学习决策函数 f(X) 或者条件概率分布 P(Y|X)作为预测的模型，即判别模型
    
    判别方法关心的是对于给定的输入 X，应该预测什么样的输出 Y


3. 生成方法特点：
    - 生成方法可以还原出联合概率分布 P(X,Y)
    - 生成方法的学习收敛速度更快
    - 但存在隐变量时，生成方法依然可用


4. 判别方法特点：
    - 判别方法可以直接学习条件概率或者决策函数
    - 往往学习的准确率更高
    - 直接学习，对数据进行抽象、定义特征并使用特征，可以简化学习
    

## 1.8 分类问题

- 当输出变量 Y 取有限个离散值时，就是一个分类问题；学习的模型称为分类器 **classifier**；分类器对新输入进行输出的预测 **prediction** 称为分类 **classification**；可能的输出称为类 **class**

- 按照分类的类别个数，划分为二分类问题和多分类问题；分类器评价指标一般为分类准确率 **accuracy** = 给定的测试集，分类器正确分类的样本数与总样本数的比值

- 二分类的评估指标通过混淆矩阵 **Confusion Matrix** 来定义，通常将**关注的类**称为正类**Positive**，其他类称为负类**Negative**；分类器在测试集上的预测正确**True**和不正确**False**
    
    TP：将正类预测为正类的数量
    
    FN：将正类预测为负类的数量
    
    FP：将负类预测为正类的数量
    
    TN：将负类预测为负类的数量
    
    精确率 **Precision** $P = \frac{TP}{TP + FP}$
    
    召回率 **Recall** $R = \frac{TP}{TP + FN}$
    
    $F_1$ score，权衡精确率和召回率，使用调和平均数思想即可：$\frac{2}{F_1} = \frac{1}{P} + \frac{1}{R}$
    
    $F_1 = \frac{2TP}{2TP + FP + FN}$
    
- 统计学习方法用于分类：K 近邻、感知机、朴素贝叶斯、决策树、决策列表、逻辑斯蒂回归模型、支持向量机、提升方法、贝叶斯网络、神经网络、Winnow


- 分类问题实际应用：
    1. 银行业务，构建客户分类模型，对用户按照贷款风险大小进行分类
    
    2. 网络安全领域，利用日志数据的分类对非法入侵检测
    
    3. 图像处理，分类用来检测图像中是否有人脸
    
    4. 手写识别，分类用来识别手写的数字
    
    5. 互联网搜索，网页的分类有助于网页的抓取、索引和排序
    
    6. 文本分类：文本内容(政治、经济、体育)；文本特点(正面积极、反面消极)
    

## 1.9 标注问题

- 标注 **tagging** 问题可以认为是分类问题的推广，也是更复杂的结构预测 **structure prediction** 问题的简单形式

- 标注问题的输入是一个观测序列，输出是一个标记序列或者状态序列

- **Notes**：可能的标记个数是有限的，但其组合所成的标记序列的个数是依序列长度呈指数级增长的

- 标记问题的评价指标与分类问题一样，统计学习方法用于标注：隐马尔可夫模型、条件随机场

- 标记问题在信息抽取、自然语言处理 **Natural Language Processing, NLP** 领域被广泛应用 (词性标注 part of speech tagging)

- 训练数据集 $T = \{(x_1,y_1), (x_2,y_2), \dots, (x_i,y_i), \dots, (x_N,y_N)\}$
    
  其中 $ x_i = (x_i^{(1)}, x_i^{(2)}, \dots, x_i^{(n)})^T , i = 1, 2, 3, \dots, N$ 是输入观测序列
  
  其中 $ y_i = (y_i^{(1)}, y_i^{(2)}, \dots, y_i^{(n)})^T$ 是相应的输出标记序列，n 是序列的长度
  
- 学习系统基于训练数据集构建一个模型，表示为条件概率分布：$P(Y^{(1)}, Y^{(2)}, \dots, Y^{(n)} | X^{(1)}, X^{(2)}, \dots, X^{(n)})$
  
  其中每一个 $X^{(i)} (i = 1, 2, \dots, n)$ 取值为所有可能的观测
  
  其中每一个 $Y^{(i)} (i = 1, 2, \dots, n)$ 取值为所有可能的标记，一般 $n << N$
  
- 标注系统按照学习得到的条件概率分布模型，对新的输入观测序列找到相应的输出标记序列

  - 具体而言，对一个观测序列 $x_{N+1} = (x_{N+1}^{(1)}, x_{N+1}^{(2)}, \dots, x_{N+1}^{(n)})^T$ 找到使得条件概率 
  
  
  $P((y_{N+1}^{(1)}, y_{N+1}^{(2)}, \dots, y_{N+1}^{(n)})^T | (x_{N+1}^{(1)}, x_{N+1}^{(2)}, \dots, x_{N+1}^{(n)})^T $ 最大的标记序列
  
  
  $y_{N+1} = (y_{N+1}^{(1)}, y_{N+1}^{(2)}, \dots, y_{N+1}^{(n)})^T$

### 习题 1.1
&emsp;&emsp;说明伯努利模型的极大似然估计以及贝叶斯估计中的统计学习方法三要素。伯努利模型是定义在取值为0与1的随机变量上的概率分布。假设观测到伯努利模型$n$次独立的数据生成结果，其中$k$次的结果为1，这时可以用极大似然估计或贝叶斯估计来估计结果为1的概率。

**Solution：**

伯努利模型的极大似然估计以及贝叶斯估计中的**统计学习方法三要素**如下：

1. **极大似然估计**  

**模型：** $\mathcal{F}=\{f|f_p(x)=p^x(1-p)^{(1-x)}\}$  

**策略：** 最大化似然函数  

**算法：** $\displaystyle \mathop{\arg\min}_{p} L(p)= \mathop{\arg\min}_{p} \binom{n}{k}p^k(1-p)^{(n-k)}$

2. **贝叶斯估计**  

**模型：** $\mathcal{F}=\{f|f_p(x)=p^x(1-p)^{(1-x)}\}$  

**策略：** 求参数期望  

**算法：**

$$\begin{aligned}  E_\pi\big[p \big| y_1,\cdots,y_n\big]
& = {\int_0^1}p\pi (p|y_1,\cdots,y_n) dp \\
& = {\int_0^1} p\frac{f_D(y_1,\cdots,y_n|p)\pi(p)}{\int_{\Omega}f_D(y_1,\cdots,y_n|p)\pi(p)dp}dp \\
& = {\int_0^1}\frac{p^{k+1}(1-p)^{(n-k)}}{\int_0^1 p^k(1-p)^{(n-k)}dp}dp
\end{aligned}$$

**伯努利模型的极大似然估计：**  

定义$P(Y=1)$概率为$p$，可得似然函数为：$$L(p)=f_D(y_1,y_2,\cdots,y_n|\theta)=\binom{n}{k}p^k(1-p)^{(n-k)}$$方程两边同时对$p$求导，则：$$\begin{aligned}
0 & = \binom{n}{k}[kp^{k-1}(1-p)^{(n-k)}-(n-k)p^k(1-p)^{(n-k-1)}]\\
& = \binom{n}{k}[p^{(k-1)}(1-p)^{(n-k-1)}(m-kp)]
\end{aligned}$$可解出$p$的值为$p=0,p=1,p=k/n$，显然$\displaystyle P(Y=1)=p=\frac{k}{n}$  

**伯努利模型的贝叶斯估计：**  

定义$P(Y=1)$概率为$p$，$p$在$[0,1]$之间的取值是等概率的，因此先验概率密度函数$\pi(p) = 1$，可得似然函数为： $$L(p)=f_D(y_1,y_2,\cdots,y_n|\theta)=\binom{n}{k}p^k(1-p)^{(n-k)}$$  
根据似然函数和先验概率密度函数，可以求解$p$的条件概率密度函数：$$\begin{aligned}\pi(p|y_1,\cdots,y_n)&=\frac{f_D(y_1,\cdots,y_n|p)\pi(p)}{\int_{\Omega}f_D(y_1,\cdots,y_n|p)\pi(p)dp}\\
&=\frac{p^k(1-p)^{(n-k)}}{\int_0^1p^k(1-p)^{(n-k)}dp}\\
&=\frac{p^k(1-p)^{(n-k)}}{B(k+1,n-k+1)}
\end{aligned}$$所以$p$的期望为：$$\begin{aligned}
E_\pi[p|y_1,\cdots,y_n]&={\int}p\pi(p|y_1,\cdots,y_n)dp \\
& = {\int_0^1}\frac{p^{(k+1)}(1-p)^{(n-k)}}{B(k+1,n-k+1)}dp \\
& = \frac{B(k+2,n-k+1)}{B(k+1,n-k+1)}\\
& = \frac{k+1}{n+2}
\end{aligned}$$
$\therefore \displaystyle P(Y=1)=\frac{k+1}{n+2}$

### 习题 1.2
&emsp;&emsp;通过经验风险最小化推导极大似然估计。证明模型是条件概率分布，当损失函数是对数损失函数时，经验风险最小化等价于极大似然估计。

**Solution：**

假设模型的条件概率分布是$P_{\theta}(Y|X)$，现推导当损失函数是对数损失函数时，极大似然估计等价于经验风险最小化。
极大似然估计的似然函数为：$$L(\theta)=\prod_D P_{\theta}(Y|X)$$两边取对数：$$\ln L(\theta) = \sum_D \ln P_{\theta}(Y|X) \\ 
\mathop{\arg \max}_{\theta} \sum_D \ln P_{\theta}(Y|X) = \mathop{\arg \min}_{\theta} \sum_D (- \ln P_{\theta}(Y|X))$$ 
反之，经验风险最小化等价于极大似然估计，亦可通过经验风险最小化推导极大似然估计。

### 使用最小二乘法拟和曲线

高斯于1823年在误差 $e_1, \dots, e_n$  独立同分布的假定下,证明了最小二乘方法的一个最优性质: 在所有无偏的线性估计类中,最小二乘方法是其中方差最小的
对于数据 $(x_i, y_i)   (i=1, 2, 3...,m)$, 拟合出函数 $h(x)$, 存在有误差，即残差：$r_i=h(x_i)-y_i$

此时 $L2$ 范数 (残差平方和) 最小时，$h(x)$ 和 $y$ 相似度最高，更拟合

一般的 $H(x)$ 为 $n$ 次的多项式，$H(x) = w_0 + w_1x + w_2x^2 + ... w_nx^n$，其中 $w(w_0,w_1,w_2,...,w_n)$ 为参数

最小二乘法就是要找到一组 $w(w_0,w_1,w_2,...,w_n)$ ，使得 $\sum \limits_{i=1}^n(h(x_i)-y_i)^2$ (残差平方和) 最小

即，求 $min\sum \limits_{i=1}^n(h(x_i)-y_i)^2$

----
举例：我们用目标函数$y=sin2{\pi}x$, 加上一个正态分布的噪音干扰，用多项式去拟合【例1.1 11页】

```python
import numpy as np
import scipy as sp
from scipy.optimize import leastsq
import matplotlib.pyplot as plt
%matplotlib inline
```

* ps: numpy.poly1d([1,2,3])  生成  $1x^2+2x^1+3x^0$*

```python
# 目标函数
def real_func(x):
    return np.sin(2 * np.pi * x)

# 多项式
def fit_func(p, x):
    f = np.poly1d(p)
    return f(x)

# 残差
def residuals_func(p, x, y):
    ret = fit_func(p, x) - y
    return ret

# 十个点
x = np.linspace(0, 1, 10)
x_points = np.linspace(0, 1, 1000)
# 加上正态分布噪音的目标函数的值
y_ = real_func(x)
y = [np.random.normal(0, 0.1) + y1 for y1 in y_]


def fitting(M=0):
    """
    M    为 多项式的次数
    """
    # 随机初始化多项式参数
    p_init = np.random.rand(M + 1)
    # 最小二乘法
    p_lsq = leastsq(residuals_func, p_init, args=(x, y))
    print('Fitting Parameters:', p_lsq[0])

    # 可视化
    fig = plt.figure(figsize = (8,6), dpi = 90)
    plt.plot(x_points, real_func(x_points), label='real')
    plt.plot(x_points, fit_func(p_lsq[0], x_points), label='fitted curve', color="red")
    plt.plot(x, y, 'bo', label='noise')
    plt.grid(True, linestyle='-.')
    plt.legend()
    return p_lsq
```

```python
# M = 0
p_lsq_0 = fitting(M=0)
```

```python
# M = 1
p_lsq_1 = fitting(M=1)
```

```python
# M = 3
p_lsq_3 = fitting(M=3)
```

```python
# M = 9
p_lsq_9 = fitting(M=9)
```

当 M = 9 时，多项式曲线通过了每个数据点，但是造成了过拟合

### 正则化

结果显示过拟合， 引入正则化项(regularizer)，降低过拟合

$Q(x)=\sum \limits_{i=1}^n(h(x_i)-y_i)^2+\lambda||w||^2$

回归问题中，损失函数是平方损失，正则化可以是参数向量的L2范数,也可以是L1范数

- L1: regularization\*abs(p)

- L2: 0.5 \* regularization \* np.square(p)

```python
regularization = 0.0001

def residuals_func_regularization(p, x, y):
    ret = fit_func(p, x) - y
    ret = np.append(ret, np.sqrt(0.5 * regularization * np.square(p)))  # L2范数作为正则化项
    return ret
```

```python
# 最小二乘法,加正则化项
p_init = np.random.rand(9 + 1)
p_lsq_regularization = leastsq(
    residuals_func_regularization, p_init, args=(x, y))

fig = plt.figure(figsize = (8,6), dpi = 90)
plt.plot(x_points, real_func(x_points), label='real', color="blue")
plt.plot(x_points, fit_func(p_lsq_9[0], x_points), label='fitted curve', color="green")
plt.plot(x_points, fit_func(p_lsq_regularization[0], x_points), label='regularization', color="red")
plt.plot(x, y, 'bo', label='noise')
plt.grid(True, linestyle='-.')
plt.legend()
```
